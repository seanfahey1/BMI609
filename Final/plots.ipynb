```python
import plotly.graph_objects as go
from plotly.io import to_html
from plotly.subplots import make_subplots
import pandas as pd

pvalues_file = "All-adj-p-values.csv"
```

```python
with open(pvalues_file, 'r') as file:
    df = pd.read_csv(file, header=0, delimiter=',', index_col=0)
df.head()
```

```python
fig1 = go.Figure()
df_wnt3a = df.sort_values(by='wnt3a_unadjusted', ascending=True)
fig1.add_traces([
    go.Scatter(
        x = list(df_wnt3a.index),
        y = df_wnt3a['wnt3a'],
        name = '  Bonferroni<br>  adjusted<br>  p-value'
    ),
    go.Scatter(
        x = list(df_wnt3a.index),
        y = df_wnt3a['wnt3a_unadjusted'],
        name = '  unadjusted<br>  p-value',
    ),
    go.Scatter(
        x = [df_wnt3a[df_wnt3a["wnt3a_unadjusted"] >= 0.05].index[0], df_wnt3a[df_wnt3a["wnt3a_unadjusted"] >= 0.05].index[0]],
        y = [0, 1],
        name = "  unadjusted<br>  0.05 cutoff",
        mode='lines',
        line = dict(color='royalblue', width=0.5, dash='dash')
    )
]).update_layout(
    width = 600, 
    height = 450,
    title = "<b>wnt3a treatment group</b>",
    template = 'plotly_white',
    # showlegend=False,
    autosize=False,
    margin=dict(
        l=30,
        r=30,
        b=85,
        t=50,
        pad=0
    )
)
fig1.show()
```
